```python
D0 = {}  # dizionario vuoto
```

```python
D1 = {'insect': 'butterfly', 'quantity': 4, 'color': 'blue'}  # un dizionario

print(D1.keys())    # estrai le chiavi del dizionario
print(D1.values())  # estrai solo i valori
print(D1.items())   # estrai tutte le coppie chiave-valore

print(D1['insect'])    # ottieni l'insetto
print(D1['quantity'])  # ottieni la quantità
print(D1['color'])     # ottieni il colore
```

```python
# Dizionario con più elementi per chiave
D2 = {'fruit': ['apple', 'orange'], 'quantity': [3, 5], 'color': ['red', 'orange']}
print(D2['fruit'])
print(D2['quantity'])
print(D2['color'])

print(D2['fruit'][0])     # ottieni il primo elemento
print(D2['quantity'][0])  # quantità del primo elemento
print(D2['color'][0])     # colore del primo elemento
```

```python
# Dizionario dentro un dizionario

D3 = {'character': {'name': 'Luke', 'hometown': 'Tatooine'}}
print(D3['character']['name'])
print(D3['character']['hometown'])
```

```python
# Un altro modo per definire un dizionario

D4 = dict(name='Luke', hometown='Tatooine')
print(D4['name'])
print(D4['hometown'])
```

```python
# Dizionario creato con zip
D5 = dict(zip(['name', 'hometown'], ['Luke', 'Tatooine']))
print(D5['name'])
print(D5['hometown'])
```

```python
# Un altro formato
D6 = dict([('name', 'Luke'), ('hometown', 'Tatooine')])
print(D6['name'])
print(D6['hometown'])
```

```python
# Ancora un formato
D7 = {}
D7['name'] = 'Luke'
D7['hometown'] = 'Tatooine'
print(D7['name'])
print(D7['hometown'])
```

```python
# Ancora un formato
dict.fromkeys(['name', 'hometown'], ['Luke', 'Tatooine'])
```

```python
# Verifica del dizionario
D2 = {'fruit': ['apple', 'orange'], 'quantity': [3, 5], 'color': ['red', 'orange']}
print('fruit' in D2)
print('quantity' in D2)
print('color' in D2)

for key in D2:
    print(key)
```

```python
D_fruit = {'fruit': ['apple', 'orange'], 'quantity': [3, 5], 'color': ['red', 'orange']}
qual = {'shape': ['round', 'round']}

D_fruit.update(qual)  # aggiunge altre chiavi al dizionario
print(D_fruit)
```

```python
# Trasforma le chiavi in lista
D_veggies = {'potato': 1, 'celery': 2, 'carrot': 3, 'spinach': 4}
list(D_veggies.keys())
```

```python
# Pop (rimozione)
D_veggies = {'potato': 1, 'celery': 2, 'carrot': 3, 'spinach': 4}
D_veggies.pop('potato')
print(D_veggies)
D_veggies.popitem()
print(D_veggies)
```

```python
# Get
D_veggies = {'potato': 1, 'celery': 2, 'carrot': 3, 'spinach': 4}
D_veggies.get('carrot')
```

```python
## Unpacking (espansione) di dizionari
D = dict(a=7, c=2)
{'a': 99, 'b': 1, **D}  # se una chiave compare due volte, vale il valore più a destra
# dict(a=99, b=1, **D)
D  # stessa cosa, ma sintassi diversa: un modo per combinare dizionari
```

```python
# Ancora un modo

power_of_two = {x: 2 ** x for x in range(11)}
print(power_of_two)
```

```python
## Gestire valori inesistenti

matrix = {1: [1, 2, 3], 2: [40, 50, 60], 3: [700, 800, 900]}
print(matrix)

print(matrix.get(1, 0))  # se la chiave 1 esiste, stampa il valore; altrimenti stampa 0
print(matrix.get(4, 0))  # la chiave 4 non esiste, quindi viene restituito 0
```

```python
## Gestire valori inesistenti

matrix = {1: [1, 2, 3], 2: [40, 50, 60], 3: [700, 800, 900]}
print(matrix)

if 1 in matrix:  # come sopra, ma con un if/else
    print(matrix[2])
else:
    print(0)
```

```python
## Gestire valori inesistenti

matrix = {1: [1, 2, 3], 2: [40, 50, 60], 3: [700, 800, 900]}
print(matrix)

try:  # come sopra, ma con try/except
    print(matrix[4])  # la chiave 4 non esiste
except KeyError:
    print(0)
```

```python
# Usare tuple come chiavi
sales = {('New York', 'January'): 5000, ('Los Angeles', 'January'): 7000}
print(sales[('New York', 'January')])  # output: 5000
print(sales)
```
