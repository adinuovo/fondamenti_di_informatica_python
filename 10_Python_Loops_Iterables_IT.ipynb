# Cicli while

```python
## NON ESEGUIRE QUESTO!
## I cicli while continueranno a girare finché la condizione è vera.
## L'esempio sotto è chiamato *ciclo infinito*: continuerà a girare perché
## non c'è nulla che cambi la condizione, quindi sarà sempre vera.
## Se per sbaglio crei un ciclo infinito, premi Ctrl+C o clicca il pulsante Stop.

# while True:
#     print('Control+C')
```

```python
x = 0
while x < 5:  # Questo gira solo se x è minore di 5, ma x cambia nel ciclo
    print(x)  # Tutto ciò che è nel ciclo deve essere indentato
    x += 1  # scorciatoia per x = x + 1: aggiunge 1 a x a ogni iterazione

print('x now equals ' + str(x))  # Fuori dal ciclo: il while si ferma perché x=5 rende x<5 False
```

```python
while False:  # Il ciclo non gira perché la condizione è falsa
    print("This won't run")
else:  # Puoi aggiungere un else che viene eseguito quando il ciclo termina
    print("This will run")
```

```python
## Possiamo riscrivere l'esempio così:
x = 0
while x < 5:  # Questo gira solo se x è minore di 5, ma x cambia nel ciclo
    print(x)  # Tutto ciò che è nel ciclo deve essere indentato
    x += 1  # scorciatoia per x = x + 1
else:
    print('x now equals ' + str(x))  # Ora è dentro l'else: il while finisce e passa qui
```

```python
# Uso dell'istruzione pass

count = 0

while count < 5:
    if count == 2:
        # Non fare nulla quando count vale 2
        pass
    else:
        print('Count is:', count)
    count += 1
```

```python
# Uso di ...

count = 0

while count < 5:
    if count == 2:
        # Non fare nulla quando count vale 2
        ...  # Fa la stessa cosa di pass
    else:
        print('Count is:', count)
    count += 1
```

```python
# continue: salta all'inizio del ciclo
# Usalo con parsimonia.
count = 0

while count < 5:
    count += 1
    if count == 3:
        # Salta il resto del ciclo quando count vale 3
        continue
    print('Count is:', count)  # Questo viene saltato quando count == 3
```

```python
# break
x = 0
while True:  # Normalmente sarebbe un ciclo infinito, ma break lo ferma
    x += 1  # Aggiunge 1 a x a ogni iterazione
    print(x)
    if x == 5:
        break  # Ferma il ciclo
```

```python
# pass, ..., continue e break funzionano anche nei cicli for.
```

## Cicli interattivi

```python
## Cicli interattivi

while True:
    answer = input('Enter text:')  # input mette in pausa il codice e aspetta l'utente
    if answer == 'stop':  # questo ciclo continua finché l'utente non scrive stop
        break  # interrompe il ciclo
    print(f'Received: {answer}')  # Fuori dall'if: esegue finché answer != 'stop'
```

```python
## Un altro esempio

while True:
    answer = input('Enter number:')
    if answer == 'stop':
        break
    multiply = int(answer) * 5  # Devi convertirlo in intero, altrimenti non puoi fare operazioni matematiche
    print(f'Multiplied by 5= {multiply}')
```

```python
## Aggiungi avvisi di errore così il codice può continuare a essere eseguito.
# Il problema del codice sopra è che, se inserisci qualcosa che non è un numero,
# otterrai un errore invece del risultato atteso. Con elif puoi dire all'utente
# che l'input non è una cifra.

while True:
    answer = input('Enter number:')
    if answer == 'stop':
        break
    elif not answer.isdigit():
        print(f'{answer} is not a number')  # Stampa e continua, così l'utente può riprovare
    else:
        multiply = int(answer) * 5
        print(f'Multiplied by 5= {multiply}')
```

```python
## Funziona allo stesso modo, ma con un try
while True:
    answer = input('Enter number:')
    if answer == 'stop':
        break
    try:  # Prova a eseguire questo blocco
        multiply = int(answer) * 5
        print(f'Multiplied by 5= {multiply}')
    except:  # Se c'è un errore, esegui questo
        print(f'{answer} is not a number')
```

# Cicli for

I cicli for sono cicli iterativi con un numero di iterazioni determinato, a differenza dei cicli while.

```python
x = [1, 3, 3, 8, 5]  # una lista

for i in x:  # itera sulla lista x (i è solo un contatore/variabile qualunque)
    print(i)  # stampa i
```

```python
fruits = ['apple', 'cherry', 'orange', 'banana', 'pear', 'watermelon']
for fruit in fruits:  # ci sono sei frutti: il ciclo prende ognuno (in ordine) e lo chiama fruit
    print('I ate a(n) ' + fruit)  # stampa la frase con il frutto corrente
```

```python
sum_num = 0  # devi creare sum_num prima del ciclo perché viene usata nel ciclo
for num in [1, 2, 3, 4]:
    sum_num = sum_num + num  # aggiunge il numero successivo
    print(sum_num)  # è nel ciclo ed esegue a ogni iterazione
print('The final answer is:', sum_num)  # fuori dal ciclo: eseguita per ultima

# un modo più veloce è:
# print(sum([1,2,3,4]))
```

```python
matrix = [[1, 2],
          [3, 4],
          [5, 6]]

for a, b in matrix:  # puoi fare unpacking usando due variabili
    print(a, b, a + b)
```

```python
# I cicli for funzionano anche per i dizionari

D = {'First Name': 'Luke', 'Last Name': 'Skywalker', 'Hometown': 'Tatooine'}
for key in D:
    print(key + ': ' + D[key])
```

```python
## Cicli for annidati
x = [1, 2, 3]
y = ['a', 'b', 'c']

for i in x:
    for j in y:
        print(i, j)

# I cicli annidati possono essere utili, ma prova a trovare un altro modo di risolvere il problema.
# I cicli annidati possono essere molto lenti a seconda dell'applicazione.
```

```python
## Cicli for annidati con if
x = [1, 2, 3]
y = [1, 3, 4]

for i in x:
    for j in y:
        if i == j:
            print('Equal', i, j)
        else:
            print('Not equal', i, j)
```

### range

```python
# puoi definire un intervallo di numeri per un ciclo for
for i in range(5):  # range va da 0 a n-1, dove n è il numero inserito
    print(i)
```

```python
# range aiuta nelle iterazioni: è un tipo di oggetto a sé
print(range(5))
print(list(range(5)))  # devi convertirlo in list() per ottenere una lista
```

```python
for i in range(2, 5):  # puoi scegliere un valore di partenza
    print(i)
```

```python
for i in range(2, 10, 2):  # il terzo numero è il passo: stampa solo numeri pari da 2 a 8
    print(i)
```

```python
# puoi creare liste con intervalli specifici
num_list = []  # devi definire prima la lista, anche se vuota
for i in range(0, 37, 9):
    num_list.append(i)  # append aggiunge alla lista
print(num_list)  # Questa riga è fuori dal ciclo (non indentata)

# un modo più veloce:
# list(range(0,37,9))
```

```python
# prova a non usare range in un ciclo for quando non serve: è più semplice iterare sugli elementi
items = ['candy bar', 'rubber band', 'paper clip']
for item in items:
    print(item)

# Il codice sopra è meglio di:
for i in range(len(items)):
    print(items[i])
```

```python
# Un caso in cui range serve davvero
num = [1, 2, 3, 4, 5]
for i in range(len(num)):
    num[i] = num[i] * 10  # puoi riassegnare elementi di una lista accedendo all'indice

print(num)
```

```python
# Un codice più compatto
num = [1, 2, 3, 4, 5]
num = [n * 10 for n in num]
print(num)
```

```python
# Questo NON funziona, perché stai cambiando n, non num
num = [1, 2, 3, 4, 5]
for n in num:
    n = n * 10

print(num)
```

### zip

```python
num1 = [1, 2, 3, 4]
num2 = [5, 6, 7, 8]
print(zip(num1, num2))  # questo diventa un oggetto zip
print(list(zip(num1, num2)))  # converte lo zip in lista
```

```python
# Uso di zip in un ciclo for
num1 = [1, 2, 3, 4]
num2 = [5, 6, 7, 8]
for a, b in zip(num1, num2):  # estrae in parallelo dalle due liste
    print(a + b)
```

```python
# zip non è limitato a 2 oggetti
M1 = [1, 2, 3]
M2 = [4, 5, 6]
M3 = [7, 8, 9]

list(zip(M1, M2, M3))
```

```python
# puoi fare zip di lunghezze diverse, ma si ferma alla più corta
N1 = [1, 2, 3, 4, 5]
N2 = ['a', 'b', 'c']
list(zip(N1, N2))
# usa solo fino a 3 da N1
```

```python
# zip rende più facile creare dizionari
keys = ['fruit', 'color', 'amount']
vals = ['apple', 'red', 5]

D = {}  # crea un dizionario vuoto

for (k, v) in zip(keys, vals):
    D[k] = v

print(D)
```

```python
# Un modo più compatto
{k: v for (k, v) in zip(keys, vals)}
```

### enumerate

```python
fruit = 'apple'
for (i, letter) in enumerate(fruit):  # i è la posizione e letter è il carattere nella stringa
    print(i, letter)
```

```python
# enumerate è iterabile
E = enumerate(fruit)
print(E)
print(next(E))
print(next(E))
```

```python
# Un esempio un po' contorto
num1 = [1, 2, 3, 4]
num2 = [5, 6, 7, 8]
for (i, n1) in enumerate(num1):
    print(i, n1 + num2[i])  # somma gli elementi nella stessa posizione come prima
```

### Mettiamo tutto insieme

```python
for ii in enumerate(zip(range(1, 5), ['a', 'b', 'c', 'd'])):
    print(ii)
```

# map e filter

```python
## map può "mappare" funzioni su altri oggetti in modo iterativo

# Trovare il massimo
L = [1, 2, 3, 4]
max_L = max(L)
print(max_L)

M = [[1, 2, 4, 1],
     [5, 9, 4, 3],
     [25, 10, 4, 9],
     [20, 36, 49, 87]]

max_M = map(max, M)  # map prende la funzione e poi l'oggetto su cui agisce
print(max_M)  # questo è solo l'oggetto map
print(list(max_M))  # devi convertirlo in lista per vedere i risultati

# altrimenti serve un ciclo for
max_M = []
for i in M:
    max_M.append(max(i))
print(max_M)
```

# List comprehension

Le abbiamo già viste, ma entriamo più nel dettaglio: spesso sono più veloci di un ciclo for standard.

```python
# Esempio
num = [1, 2, 3, 4, 5]
square = [x**2 for x in num]  # È quasi un for "al contrario": prima il risultato, poi su cosa iterare
square

## Versione con for
# square = []
# for x in num:
#     square.append(x**2)
```

```python
# Questo funziona molto bene quando si usano file.
# Per prima cosa creiamo un file.
with open('hello.txt', 'w') as file:
    file.write('Hello World!\n')
    file.write("What's up Mars!\n")
    file.write('Hi Venus!\n')
# Ora c'è un file nella directory di lavoro corrente.
# Ora leggiamolo e mettiamolo in una lista.
[line for line in open('hello.txt')]
# Nota: questa notazione chiude automaticamente il file, come con 'with open() as file:'
```

```python
# Ci sono simboli di a capo. Possiamo rimuoverli mentre li carichiamo.
[line.strip() for line in open('hello.txt')]
```

```python
# Possiamo concatenare comandi per trasformare ulteriormente l'output.
[line.strip().upper().replace('!', '?').split() for line in open('hello.txt')]
# strip rimuove i caratteri di a capo
# upper rende tutto maiuscolo
# replace cambia ogni '!' in '?'
# split crea una lista spezzando la stringa sugli spazi
# Invece di scrivere una riga per ogni operazione, puoi concatenarle come sopra
```

```python
# Questa versione aggiunge un if: carica solo le righe che iniziano con la lettera H.
[line.strip() for line in open('hello.txt') if line.startswith('H')]
```

```python
# Un codice ancora più corto
list(open('hello.txt'))
```
