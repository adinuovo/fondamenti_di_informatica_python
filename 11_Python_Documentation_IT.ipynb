# Funzione dir

```python
# Questo elenca tutti gli attributi di una lista, ma alcuni non ci servono
# perché vengono gestiti in background. Quelli hanno spesso uno o due underscore
# nel nome. Prova a de-commentare e a eseguire per vedere.
# dir(list)
```

```python
# Filtriamo gli attributi che iniziano con underscore.
[att for att in dir(list) if not att.startswith('__')]
```

```python
# Te li ricordi? Abbiamo usato append spesso nella lezione sui cicli e
# abbiamo visto gli altri nella lezione sulle liste.
```

```python
# Facciamo lo stesso con le stringhe
[att for att in dir(str) if not att.startswith('__')][:5]
# Ci sono molte altre cose che puoi fare e che non abbiamo ancora visto.
# Qui sopra ci sono i primi 5.
```

```python
# Possiamo farlo anche per i moduli.
import math
[att for att in dir(math) if not att.startswith('__')][:5]
# Qui ne ho mostrati solo 5 per farti vedere alcuni dei tanti attributi di math.
# Rimuovi '[:5]' per vederli tutti.
```

# Docstring

```python
# Le docstring stanno nel corpo di una funzione e aiutano a chiarire lo scopo
# della funzione. Studieremo le funzioni nella prossima lezione. Per ora,
# ecco un esempio.

def add_numbers(a: int, b: int) -> int:
    """
    Somma due numeri e restituisce il risultato.

    Parametri:
        a (int): il primo numero.
        b (int): il secondo numero.

    Restituisce:
        int: la somma di a e b.
    """
    return a + b

## La docstring è tra triple virgolette e, come detto prima, può funzionare come
# un commento "elegante" (se usata correttamente).
```

```python
help(add_numbers)  # Se usiamo la funzione help, viene mostrata la docstring.

# Tienilo a mente quando inizierai a scrivere le tue funzioni, così saranno
# più accessibili per altre persone che potrebbero usarle in futuro.
```

# Errori comuni

1. Non dimenticare i due punti (:). Servono in if, cicli e funzioni.
2. Usa TAB per indentare (ma parti sempre dall'estrema sinistra per il codice di livello superiore).
3. Prova a usare semplici cicli for invece di while o range quando possibile.
4. Fai attenzione al tipo di oggetto: alcuni sono mutabili e altri no; alcuni mantengono l'ordine, altri no.
5. Attenzione alle operazioni di modifica *in place* (es.: usa `L.append(...)` e non `L = L.append(...)`).
6. Usa le parentesi per le funzioni (es.: `file.close()`).
7. Non usare l'estensione nel comando import.
8. Se riassegni una lista, le variabili possono restare collegate (vedi esempio sotto).

```python
L = [1, 2, 3]
M = L
N = L.copy()
L.append(4)
print(L)
print(M)  # M non è stata aggiornata direttamente, ma è collegata a L, quindi stampa lo stesso contenuto
print(N)  # N è una copia, quindi non è collegata a L
```
