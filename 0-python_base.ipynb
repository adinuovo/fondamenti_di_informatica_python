### Istruzione If

Forse il tipo di istruzione condizionale più noto è l'istruzione `if`. Ad esempio:

```python
if x < 0 :
    print('x =', x, 'is negative')
elif x == 0:
    print('x =', x, 'is zero')
else:
    print('x =', x, 'is positive')
```

**D:** Dai un valore alla variabile `x` e vedi cosa fa questa istruzione.

**Importante!** La particolarità principale della sintassi Python è che lo scope delle diverse strutture (funzioni, for, if, while, ecc...) è definito dall'indentazione, non dalle parentesi graffe `{}`. Finché il codice rimane allo stesso livello, è nello stesso scope:

```python
if x < 0 :
    print('x =', x, 'is negative')
    x = -x
    print('x =', x, 'is now positive')
elif x == 0:
    print('x =', x, 'is zero')
else:
    print('x =', x, 'is positive')
```

Una scelta ragionevole è usare quattro spazi per l'indentazione invece delle tabulazioni (configura il tuo editor di testo se non stai usando Jupyter).

Quando lo scope è terminato, devi tornare **esattamente** allo stesso livello di indentazione. Prova questa struttura disallineata e osserva cosa succede:

```python

    print('x =', x, 'is negative')
 elif x == 0:
    print('x =', x, 'is zero')
 else:
    print('x =', x, 'is positive')
```

Jupyter è abbastanza gentile da evidenziarlo per te, ma non tutti gli editor di testo lo fanno...

In un'istruzione if, possono esserci zero o più parti elif. Cosa fare quando la condizione è vera deve essere indentato. La parola chiave `"elif"` è una forma abbreviata di `"else if"`, ed è utile per evitare un'eccessiva indentazione. Una sequenza `if ... elif ... elif ...` è un sostituto delle istruzioni switch o case presenti in altri linguaggi.

Le istruzioni `elif` ed `else` sono opzionali. Puoi anche usare solo l'istruzione if da sola:


### Classi

Le classi sono strutture che permettono di:

1. Memorizzare informazioni in un oggetto.
2. Applicare funzioni su queste informazioni.

In breve:

```python
class Foo:
    
    def __init__(self, val):
        self.val = val
        
    def print(self):
        print(self.val)
   
    def set_val(self, val):
        self.val = val
        self.print()
        
a = Foo(42)
a.print()
```

Questo definisce la **classe** `Foo`. Il primo metodo (obbligatorio) della classe è il **costruttore** `__init__`. Questo determina come l'**istanza** `a` sarà istanziata dopo la chiamata a `a = Foo(42)`. Il primo argomento è `self`, che rappresenta l'istanza corrente della classe. Possiamo specificare altri argomenti al costruttore (qui `val`), che possono essere elaborati o memorizzati.

Qui memorizziamo `val` come un **attributo** della classe `Foo` con `self.val`. Sono dati che saranno specifici per ogni oggetto creato: se crei `b = Foo("deep learning")`, l'attributo `self.val` avrà valori diversi tra le due istanze. Come sempre in Python, il tipo non importa, può essere un float, una stringa, un array numpy, un altro oggetto...

Gli attributi sono accessibili da ogni oggetto come:

```python
x = a.val
```

Puoi impostare il suo valore con:

```python
a.val = 12
```

Le classi possono definire **metodi** che possono manipolare gli attributi di classe come qualsiasi funzione normale. Il primo argomento **deve** sempre essere `self`. Con l'oggetto `self`, puoi accedere a tutti gli attributi (o altri metodi) dell'istanza.

Con la nostra classe giocattolo, `a.set_val(34)` fa esattamente lo stesso di `a.val = 34`, o `a.print()` è lo stesso di `print(a.val)`.

*Per esperti C++/Java:* gli attributi e i metodi sono sempre pubblici in Python. Se vuoi rendere privato un attributo, anteponi al suo nome un trattino basso, es. `self._val`. Non farà quindi parte dell'API della classe (ma può essere letto o scritto pubblicamente comunque...).

**D:** Gioca con questa classe base, crea diversi oggetti con attributi diversi, stampali, modificali, ecc.

Un concetto importante nella **programmazione orientata agli oggetti** (OOP) è l'**ereditarietà delle classi**. Non la useremo molto in questi esercizi, ma parliamone brevemente.

Ereditare una classe significa creare una nuova classe che eredita dagli attributi e dai metodi di un'altra classe (una sorta di "copia" della definizione della classe). Puoi quindi aggiungere nuovi attributi o metodi, o sovraccaricare quelli esistenti.

Esempio:

```python
class Bar(Foo):
    def add(self, val):
        self.val += val
    def print(self):
        print("val =", self.val)
```

`Bar` è una classe figlia di `Foo`. Eredita tutti gli attributi e i metodi, inclusi `__init__`, `print` e `set_val`. Crea un nuovo metodo `add` e **sovraccarica** `print`: la vecchia definizione di `print` in `Foo` non esiste più per le istanze della classe `Bar` (ma esiste per le istanze della classe `Foo`). Anche il costruttore può essere sovraccaricato, ad esempio per aggiungere nuovi argomenti:

```python
class Bar(Foo):
    def __init__(self, val, val2):
        self.val2 = val2
        super().__init__(val)
    def add(self, val):
        self.val += val
    def print(self):
        print("val =", self.val)
```

`super().__init__(val)` chiama il costruttore della classe `Foo` (la "super" classe di `bar`), quindi imposta il valore di `self.val`.
