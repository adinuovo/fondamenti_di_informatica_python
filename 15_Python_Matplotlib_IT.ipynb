## 2. Primo grafico (line plot)

Creiamo un semplice vettore `t` e la sinusoide `s`, poi tracciamo la curva.

```python
# Dati
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2 * np.pi * t)

# Figura + grafico
fig = plt.figure()
plt.plot(t, s)
plt.title("Sinusoide")
plt.xlabel("t")
plt.ylabel("sin(2πt)")
plt.show()
```

## 3. Interfaccia orientata agli oggetti (Figure/Axes)

In Matplotlib spesso è utile lavorare in modo **OO (object-oriented)**: 
- `fig` è il contenitore della figura
- `ax` (Axes) è lo spazio in cui disegniamo

```python
fig = plt.figure()
ax = plt.axes()

t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2 * np.pi * t)

ax.plot(t, s)
ax.set(xlabel="tempo (s)", ylabel="Ampiezza", title="Serie temporale: sin(2πt)")
ax.grid(True)
plt.show()
```

## 4. Seno e coseno: colori, stili di linea, legenda

### Modi comuni per specificare un colore
- Per **nome**: `'red'`, `'green'`, ... (lista: https://matplotlib.org/stable/gallery/color/named_colors.html)
- Codici brevi: `'r'`, `'g'`, `'b'`, `'c'`, `'m'`, `'y'`, `'k'` (nero)
- Scala di grigi: stringa `'0.0'` (nero) → `'1.0'` (bianco)
- Esadecimale: `'#RRGGBB'`
- Tupla RGB normalizzata: `(1.0, 0.5, 0.3)`

### Stili di linea
- `'-'` (continua), `'--'` (tratteggiata), `'-.'` (tratto-punto), `':'` (puntinata)

```python
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2 * np.pi * t)
c = np.cos(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s, color='0', linestyle='--', label='sin(t)')   # nero tratteggiato
ax.plot(t, c, color='red', linestyle='-.', label='cos(t)') # rosso tratto-punto

ax.set(title='Seno e coseno', xlabel='t', ylabel='ampiezza')
ax.legend()
ax.grid(True)
plt.show()
```

## 5. Limiti degli assi e griglia

Puoi controllare l’area mostrata impostando `xlim`, `ylim` e aggiungere la griglia.

```python
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2 * np.pi * t)
c = np.cos(2 * np.pi * t)

plt.figure()
plt.plot(t, s, label='sin(t)')
plt.plot(t, c, label='cos(t)')

plt.xlim(0.5, 1.75)
plt.ylim(-1.5, 1.5)
plt.grid(True)
plt.legend()
plt.title('Limiti assi + griglia')
plt.show()

# Alternative spesso utili:
# plt.axis('tight')  # adatta i limiti ai dati
# plt.axis('equal')  # stessa scala su x e y
```

## 6. Stile MATLAB vs Stile OO (riassunto)

- **Stile MATLAB**: funzioni `plt.*` (es. `plt.xlabel(...)`)
- **Stile OO**: metodi di `ax` (es. `ax.set_xlabel(...)`)

Esempio OO pulito usando `ax.set(...)` in un’unica chiamata.

```python
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)
ax.set(xlabel='tempo (s)', ylabel='Ampiezza', title='Serie temporale (OO)')
ax.grid(True)
plt.show()
```

## 7. Grafici a dispersione (scatter)

### 7.1 Usando `plt.plot(..., 'o')`
Con `'o'` disegni punti (marker) invece di una linea continua.

```python
t = np.arange(0.0, 2.0, 0.05)  # passo più grande per vedere meglio i punti
s = np.sin(2 * np.pi * t)

plt.figure()
plt.plot(t, s, 'o')
plt.title("Scatter con plt.plot()")
plt.xlabel("t")
plt.ylabel("sin(2πt)")
plt.show()
```

### Marker: riferimento

Vedi la tabella marker: 
https://matplotlib.org/stable/gallery/lines_bars_and_markers/marker_reference.html

Esempi:
- `'-ok'` = linea + marker cerchio + colore nero
- Personalizzazione: `markersize`, `markerfacecolor`, `markeredgecolor`, `markeredgewidth`

```python
t = np.linspace(0, 2*np.pi, 20)
s = np.sin(t)

plt.figure()
plt.plot(t, s, '-ok', label="linea nera con cerchi")

plt.plot(
    t, s, '-p', color='green',
    markersize=12,
    markerfacecolor='pink',
    markeredgecolor='black',
    markeredgewidth=2,
    label="pentagoni rosa con bordo nero"
)

plt.title("Esempi di marker")
plt.xlabel("t")
plt.ylabel("sin(t)")
plt.legend()
plt.grid(True)
plt.show()
```

### 7.2 Usando `plt.scatter()`

`plt.scatter()` è comodo per variare **colore** e **dimensione** dei punti.
> Nota: per dataset molto grandi, spesso `plt.plot()` è più veloce di `plt.scatter()`.

```python
t = np.arange(0.0, 2.0, 0.05)
s = np.sin(2 * np.pi * t)

plt.figure()
plt.scatter(t, s)
plt.title("Scatter con plt.scatter() (default)")
plt.show()

plt.figure()
plt.scatter(t, s, marker='v', color='r')
plt.title("Scatter con marker triangolo rosso")
plt.show()
```

#### Dimensione variabile dei punti

```python
x = np.random.randint(1, 11, 10)
y = np.random.randint(1, 11, 10)
dimensioni = np.random.randint(10, 1000, 10)

plt.figure()
plt.scatter(x, y, s=dimensioni)
plt.title("Scatter con dimensione variabile")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## 8. Barre d’errore con `plt.errorbar()` + banda con `fill_between()`

Esempio: sinusoide con barre d’errore costanti e area di incertezza.

```python
t = np.arange(0.0, 2.0, 0.05)
s = np.sin(2 * np.pi * t)

plt.figure()
plt.errorbar(
    t, s, yerr=0.2, fmt='o',
    color='red',      # colore dei punti
    ecolor='blue',    # colore barre d'errore
    elinewidth=3,     # spessore barre
    capsize=3         # dimensione cappucci
)

# Banda di errore (±0.2)
plt.fill_between(t, s-0.2, s+0.2, color='gray', alpha=0.3)

plt.title("Seno con barre d'errore")
plt.xlabel("t")
plt.ylabel("sin(2πt)")
plt.grid(True)
plt.show()
```

## 9. Colormap (cenno)

Le **colormap** definiscono come mappare un valore numerico in un colore.
- Panoramica: https://matplotlib.org/stable/users/explain/colors/colormaps.html

Esempio rapido: colorare i punti in base a una variabile `z`.

```python
# Esempio semplice di colormap
np.random.seed(0)
x = np.random.randn(200)
y = np.random.randn(200)
z = np.sqrt(x**2 + y**2)  # distanza dall'origine

plt.figure()
sc = plt.scatter(x, y, c=z, cmap='viridis', alpha=0.8)
plt.colorbar(sc, label='z (distanza)')
plt.title('Scatter colorato con colormap')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## 10. (Opzionale) Scatter con dataset Iris

Questa cella usa `scikit-learn`. Se non è installato nel tuo ambiente:
- in Colab di solito è già disponibile
- in locale puoi installarlo con `pip install scikit-learn`

> Se non ti interessa, puoi saltare questa sezione.

```python
# Opzionale: dataset Iris
try:
    from sklearn.datasets import load_iris
    iris = load_iris()

    attributi = iris.data.T

    fig, ax = plt.subplots()
    sc = ax.scatter(
        attributi[1], attributi[2],
        c=iris.target, cmap='cool', alpha=0.3,
        s=100 * attributi[3]
    )
    ax.set_xlabel(iris.feature_names[1])
    ax.set_ylabel(iris.feature_names[2])
    ax.set_title("Iris: scatter (colore=specie, dimensione=petal width)")
    plt.show()
except Exception as e:
    print("Cella opzionale non eseguita. Dettagli:", e)
```
