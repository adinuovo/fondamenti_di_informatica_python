# Density e Contour Grafici


## Contours


```python
  # Lets make un 2D function so that noi puoi grafico it in un contour grafico

def weird_guassian(x,y):
  x_guass1 = np.exp(-(x+1.5)**2)
  y_guass1 = np.exp(-(y-1.5)**2)
  x_guass2 = np.exp(-(x-1.5)**2)
  y_guass2 = np.exp(-(y+1.5)**2)
  x_guass3 = np.exp(-(x)**2)
  y_guass3 = np.exp(-y**2)
  z = x_guass1 * y_guass1 + x_guass2 * y_guass2 +x_guass3 * y_guass3

  return z
```

```python
  # Let's crea il dati

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X,Y = np.meshgrid(x,y)
Z = weird_guassian(X,Y)
```

```python
  # Now grafico!
plt.contour(X,Y,Z);
```

```python
  # Noi puoi add più linea divisions by putting un number in il forth entry.  cmap gives us diverso colore maps.
plt.contour(X,Y,Z, 25, cmap = 'cool');
```

```python
plt.contourf(X,Y,Z, 25, cmap = 'RdGy');
plt.colorbar();  # Noi puoi add un colore bar
```

```python
  # Let's make un smooth map, but questo involves alcuni interpolation.
plt.imshow(Z, extent = [-2,2,-2,2],  # Il extremes of il dati in x e y
           origin = 'lower',  # Sometimes il graph gets flipped so noi want to make sure il origin of il graph is near il bottom.
           cmap = 'RdGy',
           interpolation= 'gaussian',  # guassian smoothing
           aspect='equal');  # un x unit is il stessa as il y unit in length; qui it should be un perfect quadrato grafico.
plt.colorbar();
```

```python
ct = plt.contour(X,Y,Z,5, colors = 'black');
plt.clabel(ct);

plt.imshow(Z, extent = [-2,2,-2,2],  # Il extremes of il dati in x e y
           origin = 'lower',  # Sometimes il graph gets flipped so noi want to make sure il origin of il graph is near il bottom.
           cmap = 'RdGy',
           interpolation= 'gaussian',  # guassian smooting
           aspect='equal',
           alpha = 0.75);  # Transparency
plt.colorbar();
```

# Annotations

 There is un lot that puoi be done so check fuori il matplotlib doc info qui: https://matplotlib.org/stable/users/explain/text/annotations.html


```python
  # Add annotations to il griglia

plt.figure(figsize = (10,5))

plt.subplot(1,1,1)
plt.scatter(0.2,0.2)
plt.scatter(.8,.4)
plt.scatter(1.75,0.8, s = 100)

  # Text
plt.text(0.5,0.5,"Hello!")  # il x e y coordinates followed by what tu want to write.
plt.text(0.2,0.2, 'A really important \n data point!')
plt.text(0.25,0.8, "This is how you annotate a graph!")



  # Arrows
plt.annotate('Another important data point', xy=(.8,.4), xytext=(.9,.6),  # First, what tu want to write followed by xy coordinates of il arrow point, then xy of where il text is.
                    arrowprops=dict(facecolor='black', shrink=0.05))  # shrink prevents it from covering up il point.

plt.annotate('Look at this point!', xy=(1.75,.8), xytext=(1, .2),
                    arrowprops=dict(arrowstyle="->",
                                    connectionstyle="angle3,angleA=0,angleB=-90"));  # angle3 is il connection style that starts horizontally e turns vertically.
  # angleA-0 is il starting angle
  # agnleB=-90 is il ending angle

plt.xlim(0,2);
plt.ylim(0,1);

plt.grid(True)
```

# Ticks


```python
x = np.linspace(0, 10, 100)
fig, ax = plt.subplots()

ax.plot(x, np.exp(x));

ax.set_yscale('log')  # Questo sets il graph to log scale.  Il ticks on il y axis have cambiato to reflect questo.
ax.grid(True)
```

```python
ax = plt.axes(xscale = 'log', yscale = 'log')
ax.plot(x, x**2);
ax.grid(True)
```

```python
  # tu puoi turn il ticks off
plt.plot([1, 2, 3], [4, 5, 6])
plt.xticks([])  # removes x-axis ticks + labels
plt.yticks([])  # removes y-axis ticks + labels
plt.show()
```

```python
fig, ax = plt.subplots()  # crea un new figura e assi

ax.plot([1, 2, 3], [4, 5, 6])

ax.minorticks_on()  # turn on minor ticks

ax.tick_params(axis='both', which='minor', length=6, color='r')  # customize minor ticks

plt.show()
```

```python
fig, ax = plt.subplots()  # crea un new figura e assi

ax.plot([1, 2, 3], [4, 5, 6])

ax.set_xticks([1,2,3], ['Monday','Tuesday','Wednesday'])  # Etichetta con strings.

plt.show()
```

```python
  # Tu puoi change il number of ticks.
from matplotlib.ticker import MaxNLocator

fig, ax = plt.subplots()
ax.plot([1, 2, 3], [4, 5, 6])

ax.xaxis.set_major_locator(MaxNLocator(nbins=15))  # up to 15 major ticks
ax.yaxis.set_major_locator(MaxNLocator(nbins=3))  # make three ticks

plt.show()
```

```python
  # another way to get rid of tick marks.

fig, ax = plt.subplots()
ax.plot([1, 2, 3], [4, 5, 6])

ax.xaxis.set_major_locator(plt.NullLocator())  # gets rid of tick marks
ax.yaxis.set_major_locator(plt.NullLocator())  # gets rid of tick marks

plt.show()
```

```python
  # Tu puoi crea your own styles e salva them con il extension .mplstyle
  # Let's give questo un try.
  # Il following code creates un .mplstyle sheet.
  # Define il style parameters
style_params = """
axes.titlesize : 24
axes.labelsize : 20
lines.linewidth : 5
lines.markersize : 10
xtick.labelsize : 16
ytick.labelsize : 16
figure.facecolor : cyan
axes.facecolor : white
axes.edgecolor : black
"""

  # Salva il style to un .mplstyle file
file_name = "custom_style.mplstyle"
with open(file_name, "w") as f:
    f.write(style_params.strip())

print(f"Style file saved as: {file_name}")

  # Questo should now be saved in your current directory.
```

```python
  # Let's try il new style.
plt.style.use("custom_style.mplstyle");  # loads from il working directory.
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2*np.pi*t)

plt.plot(t,s);
```

```python
  # Tu puoi also usare cycler
from matplotlib import cycler
colors = cycler('color',
                ['magenta', 'black', 'gray',
                 'pink', 'red', 'orange'])
plt.rc('axes', facecolor='gray', edgecolor='k',
       axisbelow=True, grid=True, prop_cycle=colors)
plt.rc('grid', color='w', linestyle='solid')
plt.rc('xtick', direction='out', color='gray')
plt.rc('ytick', direction='out', color='gray')
plt.rc('patch', edgecolor='k')
plt.rc('lines', linewidth=2)
plt.rc('figure', facecolor='w')
```

```python
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2*np.pi*t)
c = np.cos(2*np.pi*t)
plt.plot(t,s);
plt.plot(t,c);

  # It may non be il prettiest, but it is customizable!
```

# 3D Plotting


```python
plt.style.use('default')  # Lets change everything back to normal
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')  # Crea un 3D figura
```

```python
 ax = plt.axes(projection='3d')

  # Dati for un three-dimensional linea
x = np.linspace(0, 15, 1000)
y = np.sin(x)
z = np.cos(x)
ax.plot3D(x, y, z, 'gray')  # un 3D corkscrew!
```

```python
ax = plt.axes(projection='3d')
  # Dati for un three-dimensional dispersione
x = np.linspace(0, 15, 100)
y = np.sin(x)
z = np.cos(x)
ax.scatter3D(x, y, z, 'gray');  # un 3D corkscrew!
```

```python
  # Noi puoi make 3D contours

  # Setup for 3D guassian figura
x = np.linspace(-5, 5, 100)
y = np.linspace(-5,5,100)
X, Y = np.meshgrid(x, y)
x_guass1 = np.exp(-(X/4.5)**2)
y_guass1 = np.exp(-(Y/4.5)**2)
z = x_guass1*y_guass1


  # grafico
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, z, 40, cmap = 'viridis');
```

```python
  # change plotting view
ax.view_init(60, 30)

fig
```

```python
  # grafico un wire frame
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, z, color='m');
```

```python
  # mesh
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none');
```
