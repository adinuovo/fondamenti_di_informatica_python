## Virgolette singole e doppie per le stringhe

```python
## Le virgolette singole e doppie sono sostanzialmente equivalenti
a = 'hello'
b = "world"  # ma se inizi con le doppie devi chiudere con le doppie. Lo stesso vale per le singole.
print(a, b)
```

```python
c = "hello 'world'"  # puoi comunque usare apici singoli o doppi dentro la stringa,
d = 'hello "world"'  # ma devi usare l'altro tipo per delimitare la stringa.
print(c)
print(d)
```

```python
s = "I'm a string."  # virgolette doppie con apostrofo
# Se vuoi usare un apostrofo, conviene usare le virgolette doppie all'esterno.
```

```python
# È consigliato usare in modo coerente le virgolette singole per delimitare
# le stringhe.
```

```python
## Se devi inserire un apice singolo o doppio dentro una stringa, usa \\ per fare escape

quote = "Hamlet said, \"To be or not to be that is the question.\""
print(quote)
quote2 = 'Neil Armstrong once said, "That\'s one small step for man..."'
print(quote2)
```

```python
print(s[0:5:2])   # va da 0 a 4 con passo 2
print(s[::3])     # stampa ogni terzo carattere
print(s[12:5:-1]) # ordine inverso (slice con passo negativo)
print(s[::-1])    # stringa invertita
```

## Modificare le stringhe

```python
## Le stringhe sono immutabili

s = 'string'
# s[0] = 'z'  # Questo non funziona perché le stringhe sono immutabili.
```

```python
# Puoi aggiungere o rimuovere parti
s = 'string'
s = s + 'y cheese!'
print(s)

s = s[:4] + 'pes' + s[-1]
print(s)
```

```python
## replace
s = 'stripes!'
s = s.replace('tripe', 'ponge')
print(s)
```

```python
pattern = '---&---&---&---'
print(pattern.replace('&', '$$$'))  # sostituisce tutte le occorrenze
print(pattern.replace('&', '$$$', 2))  # sostituisce le prime due occorrenze
```

```python
print(pattern)  # nota: l'oggetto non è cambiato. Cambia solo se lo riassegni.
# ad esempio
pattern = pattern.replace('&', '$$$')
print(pattern)
```

```python
## trova la posizione e sostituisci
pattern2 = '---&---&---&---'
loc = pattern2.find('&')  # trova la posizione (es. 3 per il primo &)
pattern2 = pattern2[:loc] + '$$$' + pattern2[loc+1:]
print(pattern2)

##
# pattern2 = '---&---&---&---'
# print(pattern2.replace('&', '$$$', 1))  # Fa la stessa cosa
```

```python
## sostituisci più elementi

food = 'I like apples and cake.'
food = food.replace('apples', 'bananas').replace('cake', 'pie')

print(food)
```

```python
## Converti in lista per manipolare più facilmente

greeting = 'Hello World!'
greeting_list = list(greeting)
greeting_list[6:] = 'Mars!'
greeting = ''.join(greeting_list)  # unisce gli elementi della lista senza spazi
print(greeting)
```

## Formattazione delle stringhe
3 modi diversi

Ogni modo è mostrato brevemente, ma esistono metodi più avanzati per ciascuna tecnica di formattazione.

### Espressione di formattazione con %

```python
'There are %d members of the fellowship of the %s' % (9, 'ring')
# d per decimale
# s per stringa
# i per intero
# f per float
# e per float in notazione scientifica
```

### Metodo di formattazione con format()

```python
quote = '{}, I am your {}.'  # posizione relativa
quote.format('Luke', 'father')
```

```python
quote = '%s, I am your %s.'  # posizione relativa
quote % ('Luke', 'father')
```

```python
quote = '{1[Name]}, I am your {0[family]}.'  # 0 e 1 sono la posizione in format(); posizione relativa
quote.format({'family': 'father'}, {'Name': 'Luke'})  # gli input sono dizionari
```

```python
quote_list = ['Luke', 'father']
quote = '{0[0]}, I am your {0[1]}.'
quote.format(quote_list)
```

### Formattazione con f-string
L'iterazione più recente della formattazione delle stringhe.

```python
verb = 'need'
noun = 'boat'

f'We are going to {verb} a bigger {noun}.'
```

```python
f'We are going to {verb} a bigger {noun.upper()}.'
```
