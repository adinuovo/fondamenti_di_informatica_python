## Stringhe
### Le stringhe sono sequenze che mantengono il loro ordine. Altri tipi di sequenze sono le liste e le tuple. Le stringhe, tuttavia, sono sequenze di caratteri.

```python
S = "Hello World!"
print(S)    # stampa l'intera stringa
print(S[0]) # stampa il primo carattere
print(S[1]) # stampa il secondo carattere
print(S[-1])# stampa l'ultimo carattere
print(S[-2])# stampa il penultimo carattere
```

```python
print(S[1:3]) # stampa il secondo e il terzo carattere; dall'1 al 2, ma non il 3
print(S[6:])  # stampa dalla settima posizione in avanti
print(S[:5])  # stampa dall'inizio fino alla quinta posizione,
              # ma non include la quinta posizione
print(S[:-1]) # stampa tutto tranne l'ultima posizione
print(S[:])   # stampa tutto
```

```python
## Questo esempio produce un errore, ma dimostra che non puoi aggiungere a
## una lista usando l'indicizzazione.

# LL = ['apple', 'orange', 'banana', 'peach', 'plum']
# LL[5] = 'pear'
```

```python
## Altri modi per creare un dizionario
## Occasionalmente non tornerà nello stesso ordine in cui lo digiti.
## Ma questo non importa, perché i dizionari non si curano dell'ordine.

veggies1 = dict(vegetable = 'carrot', quantity = 1, color = 'orange')               # Usando parole chiave
veggies2 = dict(zip(['vegetable', 'quantity', 'color'], ['carrot', 1, 'orange']))   # Usando la funzione zip
veggies3 = dict([('vegetable', 'carrot'), ('quantity', 1), ('color', 'orange')])    # Altra formattazione

print(veggies1)
print(veggies2)
print(veggies3)
```

```python
## Digressione sui cicli for
# i cicli for servono per iterare un processo.
# L'esempio qui sotto itera la funzione print
# basandosi su cosa è i. Nella prima iterazione i = 1
# poi i = 2, e così via fino a i = 5. Una volta che
# i = 5 è completo, il ciclo for termina.
for i in [1,2,3,4,5]:
  print(i)
```

```python
## Digressione sui cicli while
# i cicli while sono come i cicli for nel senso che iterano
# ma continueranno ad andare a meno che una certa condizione sia soddisfatta.
# Nell'esempio qui sotto, il ciclo verrà eseguito solo quando x è maggiore di 0.
# Tuttavia, all'interno del ciclo, x=x-1 sta rendendo x più piccolo di 1 ad ogni iterazione.
# Alla fine x sarà uguale a 0, e la condizione non sarà vera e il ciclo
# si fermerà.

x = 3
while x > 0:
  print(x)
  x = x-1

# Nota: evita di usare i cicli while a meno che non sia strettamente necessario. I cicli while possono girare all'infinito
# e richiederanno un arresto forzato del tuo codice. Per esempio, se cancellassi x=x-1,
# il codice continuerebbe per sempre, poiché la condizione sarebbe sempre vera (x=3),
# non importa quante iterazioni abbiano luogo.
```

```python
## Questi cicli possono essere usati su oggetti che sono iterabili come le liste.

square = [y**2 for y in [1,2,3,4,5]]
square
```

```python
square = []           # Per fare il ciclo in questo formato devi inizializzare la lista.
for y in [1,2,3,4,5]:
  square.append(y**2)
square                # Stesso risultato di sopra.
```

## Tuple
Le tuple sono liste che non possono essere cambiate.

Sono sequenze come le liste.

Ma immutabili come le stringhe.

```python
T = (1,2,3,4)
T
```

```python
print(T + (5,6))   # Concatena
print(T[0])        # Indice
```

```python
# Metodi specifici delle tuple
T2 = (3, 5, 8, 1)
print(T2.index(8))  # Il numero 8 appare alla posizione 2
print(T2.count(8))  # 8 appare una volta nella tupla
```

```python
## Le tuple sono immutabili
# T2[0] = 2
# T2.append(4)
# Questo produrrebbe un errore perché non puoi cambiare la tupla.
```

```python
T3 = 'apple', 4, [12, 13, 14]  # A volte le tuple non hanno bisogno di parentesi
T3
```

## Classi Definite dall'Utente

```python
class fruits:                               # Nome della classe
  def __init__(self, name, color, price):   # Inizializza: self (la variabile); name, color e price sono gli attributi di self
    self.name = name                        # Nome del frutto
    self.color = color                      # Colore del frutto
    self.price = price                      # Prezzo del frutto
```

```python
apple = fruits('apple', 'red', 1)
orange = fruits('orange', 'orange', 2)
banana = fruits('banana', 'yellow', 3)
```

```python
print(apple.color)
print(orange.name)
print(banana.price)
```
