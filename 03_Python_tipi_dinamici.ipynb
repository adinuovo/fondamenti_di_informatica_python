## Variabili

```python
## La tipizzazione dinamica è il processo di creazione delle variabili e di
## allocazione della memoria per tale informazione.
```

```python
# Esempio Semplice
a = 5
a
# 'a' ora punta al numero 5
```

```python
a = 'ciao'
a
# 'a' ora punta alla stringa 'ciao'. L'oggetto precedente (5) non è più referenziato.
```

### Riferimenti Condivisi
Si ha un riferimento condiviso quando più variabili puntano allo stesso oggetto in memoria.

```python
a = 5
b = a
# Sia 'a' che 'b' puntano all'oggetto intero 5.
```

```python
a = a + 2
print(a)
print(b)
# Poiché gli interi sono immutabili, 'a' punta ora a un nuovo oggetto (7),
# mentre 'b' continua a puntare a 5.
```

### Oggetti Mutabili e Riferimenti Condivisi
Con le liste (che sono mutabili), il comportamento è diverso.

```python
L1 = [2, 3, 4]
L2 = L1
L1[0] = 24
print(L1)
print(L2)
# Entrambi sono cambiati perché puntano allo stesso oggetto lista modificato in loco.
```

### Uguaglianza vs Identità
- `==` controlla se i valori sono uguali.
- `is` controlla se le variabili puntano allo stesso identico oggetto (identità).

```python
N = [1, 2, 3]
M = [1, 2, 3]
print(N == M) # I valori letterali sono uguali. Risultato: True
print(N is M) # Non puntano allo stesso oggetto. Risultato: False
```
