## Operazioni e Precedenza
La precedenza è la classifica delle operazioni che dice a Python quale operazione eseguire per prima quando ci sono più operazioni in un'espressione. Qui elencate con ## in ordine dalla precedenza più bassa a quella più alta ci sono le operazioni per un'operazione numerica. Per esempio, la moltiplicazione avverrà prima dell'addizione. Non preoccuparti troppo di queste diverse operazioni ora, ma comprendi che c'è un ordine delle operazioni. Questo può essere semplificato usando le parentesi per assicurarsi che non ci siano ambiguità.

```python
# definisci due variabili per dimostrare le operazioni
A = 1
B = 2
```

```python
## yield x, yield from x
# yield viene usato nelle funzioni, maggiori dettagli più avanti, ma ecco un esempio.
def test_function(A,B):
  yield A
  yield B

for x in test_function(A,B):
  print(x)
```

```python
## lambda args: espressione
# crea una funzione, maggiori dettagli più avanti, ma ecco un esempio.
A = 1
B = 2
add_2_num = lambda a,b: a+b
add_2_num(A,B)
```

```python
## x if y else z
# A è 1 e minore di B = 2, quindi la risposta è 400
# Se A fosse maggiore di B la risposta sarebbe 300
300 if A > B else 400
```

```python
## or
# Sarà falso solo se entrambe le affermazioni sono false, altrimenti vero.
T = True
F = False
T or F
```

```python
## and
# Sarà vero solo se entrambe le affermazioni sono vere, altrimenti falso.
T and F
```

```python
## not
# è l'opposto della variabile
print(not T)
print(not F)
```

```python
L = [1, 2, 3, 4] # crea una lista
## in
# Controlla se il valore o l'oggetto è in un altro oggetto
print("1 in L",1 in L) # controlla se 1 è in L
print("5 in L",5 in L) # controlla se 5 è in L
## not in
# Controlla se l'oggetto non è in un altro oggetto.
print("1 not in L",1 not in L) # controlla se 1 non è in L
print("5 not in L",5 not in L) # controlla se 5 non è in L
## Minore di e maggiore di (< & >)
print("1>2 è",1>2)
print("1<2 è",1<2)
## Minore o uguale a e maggiore o uguale a (<= & >=)
print("1>=2 è",1>=2)
print("1<=2 è",1<=2)
## Uguale a e non uguale a (== & !=)
print("1==2 è",1==2)
print("1!=2 è",1!=2)
```

```python
## OR Bit a bit
print(T|F)

Set1 = {1,2,3,4}
Set2 = {3,4,5,6}
## OR Bit a bit con set, dà l'unione dei set
print(Set1 | Set2)
```

```python
## XOR Bit a bit o differenza simmetrica dei set
# XOR segue questo
# 0 XOR 0 è 0
# 0 XOR 1 è 1
# 1 XOR 0 è 1
# 1 XOR 1 è 0
print(T^F)

## XOR Bit a bit con set, dà la differenza dei set
print(Set1 ^ Set2)
```

```python
## AND Bit a bit
print(T&F)

## AND Bit a bit con set, dà l'intersezione dei set
print(Set1 & Set2)
```

```python
## x<<y o x>>y sposta x di y bit

2 << 1
```

```python
## Addizione e Sottrazione
print(A+B)
print(A-B)
```

```python
## Moltiplicazione, Resto, Divisione (reale e intera/floor)

# Moltiplica
print(A*B)
# Resto della divisione, 1/2 è 0 con resto 1
print(A%B)
# Divisione (reale)
print(A/B)
# Divisione (floor, arrotonda per difetto)
print(A//B)
```

### Ordine delle Operazioni

L'ordine va dalla precedenza massima a quella minima. Le operazioni sono in ordine dalla minima alla massima nella sezione precedente (Operazioni e Precedenza). Per esempio, il codice moltiplicherà i numeri prima di aggiungerli.

```python
w = 2
x = 3
y = 4
z = 5

# Esempi di ordine delle operazioni
w*x+y*z

# 2*3+4*5
# 6+20
# 26
```

```python
# Le parentesi cambieranno l'ordine delle operazioni

w*(x+y)*z

# 2*(3+4)*5
# 2*7*5
# 70
```

### Floor vs Troncamento

```python
import math
print(3/-2)  # Mantieni il resto
print(3//-2) # Arrotonda per difetto
print(math.floor(3/-2)) # Arrotonda per difetto come sopra
print(math.trunc(3/-2)) # Arrotonda al più vicino a 0
```

## Oggetti Set
I set mostreranno un elemento solo una volta in un set, non importa quante volte venga aggiunto. Per esempio [1, 1, 3] mostrerà solo {1,3}

```python
# Due modi per creare un set
z = set('abcde')
w = {100,'j',5,'hello'}
print(z)
print(w)  # L'ordine non viene preservato.
```

```python
# Crea un set vuoto
empty_set = set()
empty_set
```

```python
not_empty_set = set(['a', 'b', 'a', 'z', 'y', 'w', 'z'])
not_empty_set
# L'output elimina i duplicati e mette le lettere in ordine.
```

```python
# Star unpacking (spacchettamento con asterisco)

{1, *'Hello', *[4,5,6]}
```

```python
A = set('abcxyz')
B = set('cdewxy')
```

```python
print(A-B)  # lettere che sono differenti; a e b esistono in A non in B
            # e z esiste in B non in A
print(A|B)  # tutte le lettere (non ripetute) in entrambi i set
print(A&B)  # tutte le lettere che sono in comune
print(A^B)  # lettere in A o B ma non in entrambi
print(A<B,A>B,A<=B,A>=B)  # test per sottoinsieme o sovrainsieme; qui tutti falsi
```

```python
C = set([1,2,3])
D = set([1,2,3,4,5])
print(C<D,C>D)  # C è un sottoinsieme di D (True); ma D non è un sottoinsieme di C (False)
```

```python
print(3 in D)  # chiedi se un elemento esiste in un dato set
print(3 not in D)  # chiedi se un elemento non esiste in un dato set
print('b' in A)    # funziona anche per le stringhe.
```

```python
F = A.intersection(B)  # uguale a A&B
print(F)

F.add('Hello')         # Inserisci elemento
print(F)

F.update(set([100, 200]))  # unisci set
print(F)

F.remove('y')              # rimuovi un elemento
print(F)
```

```python
## I set sono iterabili quindi puoi usarli nei cicli for
for num in set([1,2,3]):
  print(num*4)
```

```python
S = {5}
# S.add([1,2,3])  # Non puoi fare questo perché la lista è 'unhashable'
S.add((1,2,3))  # fai invece questo, ma lo tratta come un'unità (vedi risultato)
S
```

```python
# Per unire i set usa
S = {5}
S.update([1,2,3])
S
```

```python
# Creazione di un set con un ciclo
{x*2 for x in [1,2,3,4]}
```
