# Tuple

Le tuple sono come le liste, ma non possono essere modificate (cioè sono immutabili).

```python
T0 = ()  # tupla vuota
```

```python
T1 = (0)  # questo diventa un intero
print(T1)
T1_1 = (0,)  # questo diventa una tupla
print(T1_1)
```

```python
T2 = (0, 'Hello', 3.14, True)
print(T2)
```

```python
T3 = 0, 'Hello', 3.14, True
print(T3)
```

```python
# Tupla annidata
T4 = (0, (1, 2, 3))
print(T4)
```

```python
T5 = tuple('Hello')
print(T5)
```

```python
# Indicizzazione
T6 = (0, (1, 2, 3))
print(T6)
print(T6[0])
print(T6[1][1])
print(T6[1][1:3])
print(len(T6))
```

```python
# Concatenazione e ripetizione
T2 = (0, 'Hello', 3.14, True)
T4 = (0, (1, 2, 3))

print(T2 + T4)
print(T2 * 3)
```

```python
# Confronto

num1 = (1, 2, 3)
num2 = (1, 2, 4)
print(num1 < num2)
print(num1 == num2)
print(1 in num1)
```

```python
# Cicli
num1 = (1, 2, 3)
for i in num1:
    print(i)
```

```python
# Unpacking (come per i dizionari)

x = (1, 2, 3)
y = (4, 5, 6)

T7 = (*x, 0, *y)
print(T7)
```

```python
# index e count
T8 = (1, 5, 3, 4, 5, 5)
print(T8.index(3))  # mostra dove si trova nella tupla il numero 3
print(T8.index(5, 2))  # mostra dove si trova nella tupla il secondo 5
print(T8.count(5))  # mostra quante volte compare 5 nella tupla
```

```python
# Converti in lista
tup = (1, 2, 3)
list(tup)
```
