## Operatore walrus (:=)

Potresti incontrare un altro operatore chiamato *walrus operator*, che rende più corto il codice da scrivere.

```python
# Le istruzioni if vengono eseguite finché la condizione è vera

if True:
    print('Questo è vero')  # Questo verrà sempre stampato

if False:
    print('Questo è falso')  # Questo non verrà mai stampato
```

```python
# L'istruzione if può essere scritta in molti modi diversi a seconda di ciò che ti serve

if 2 + 2 == 4:
    print('Questo è vero')  # La riga successiva a un if DEVE essere indentata

if 100 < 200:
    print('100 è minore di 200')

txt = 'Hello World!'
if 'Hello' in txt:
    print('Hello è nel testo')

# Finché la condizione è vera, l'istruzione indentata verrà eseguita
```

```python
# E se vuoi fare qualcosa di diverso quando la condizione è falsa?

if False:
    print('Questo non viene eseguito.')
else:
    print('Successo!')

# Puoi usare il formato if/else: se la prima parte è falsa, viene eseguito il blocco sotto else
```

```python
# Altri esempi
n = 1

if n == 10:  # Questa condizione è falsa, quindi non stampa 10
    print(n)
else:  # il codice passa qui
    n = n + 1  # 1 viene aggiunto a n e assegnato alla variabile
    print(n)  # n viene stampato e dovrebbe essere 2 (se n partiva da 1)
```

```python
# Un altro esempio: lo imposto come funzione così possiamo provare più input
# (le funzioni sono trattate nella parte 10)

def use_fruit(fruit):
    if fruit == 'lemon':
        print('Fai una limonata!')
    else:
        print('Fai una macedonia!')

use_fruit('lemon')  # 'lemon' è l'input di use_fruit(fruit), poi passa nell'if/else
use_fruit('watermelon')  # 'watermelon' non è uguale a 'lemon', quindi si attiva else
use_fruit('grapes')  # idem per 'grapes'
```

```python
# E se abbiamo più percorsi possibili? Possiamo usare elif

n = 3

if n > 5:  # n è minore di 5, quindi è falso e non viene eseguito
    print('Maggiore di 5')
elif n < 5:  # n è minore di 5, quindi è vero e stamperà 'Minore di 5'
    print('Minore di 5')
else:  # Questa parte NON verrà valutata perché l'elif precedente è vero
    print('Uguale a 5')

# Prova a cambiare n per vedere le diverse risposte
```

```python
# Puoi usare quanti elif vuoi

# Altro esempio: anche qui uso una funzione così possiamo provare più input
# (le funzioni sono trattate più avanti)

def use_more_fruit(fruit):
    if fruit == 'lemon':
        print('Fai una limonata!')
    elif fruit == 'apple':
        print('Fai una torta di mele!')
    elif fruit == 'orange':
        print('Fai un succo d'arancia')
    elif fruit == 'avocado':
        print('Fai il guacamole!')
    else:
        print('Fai una macedonia!')

use_more_fruit('lemon')
use_more_fruit('avocado')
use_more_fruit('apple')
use_more_fruit('watermelon')
```

```python
# Assicurati che l'indentazione sia corretta, altrimenti possono verificarsi errori o risultati inattesi

x = 2

if x == 2:
    x = x + 2
else:
    x = 3
    x = x + 3  # questa riga è nel blocco else

print(x)  # con x=2 entra nell'if e ottieni 4
```

```python
## Ricorda i due punti (:)!!
# Le istruzioni if/elif/else richiedono i due punti: è un errore comune
```

```python
# Nota anche che solo il primo ramo con condizione True verrà eseguito: gli altri vengono ignorati.
# Quindi l'ordine delle condizioni può essere importante.

if True:
    print('Questo verrà stampato')
elif True:
    print('Questo non verrà stampato')
elif True:
    print('Nemmeno questo verrà stampato')
elif True:
    print('Nemmeno questo verrà stampato')
else:
    print('Questo non verrà stampato')
```

```python
## Puoi anche annidare istruzioni if/else

y = 10

if y > 5:
    if y < 20:
        print('y è compreso tra 5 e 20')
    else:
        print('y è maggiore di 20')
else:
    print('y è minore di 5')

# Se y è maggiore di 5 entra nell'if annidato.
# Nota che l'if/else annidato è indentato a un livello in più: è necessario per il flusso del codice.
# Se y = 2, salta l'if annidato e stampa 'y è minore di 5'. Prova numeri diversi.
```

```python
# Uso di and e or

u = 5
v = 7

if (u == 5) and (v == 6):  # entrambe le condizioni devono essere vere
    print(u + v)
else:  # dato che v non è uguale a 6, entra qui e stampa -2
    print(u - v)

if (u == 5) or (v == 6):  # basta che una condizione sia vera
    print(u + v)  # almeno u==5, quindi stampa 12
else:
    print(u - v)
```

```python
# Un'altra sintassi
# A volte per risparmiare spazio l'if può essere scritto così (espressione condizionale)
# In questo formato devono comparire sia if sia else sulla stessa riga

x = 2
y = x + 2 if x > 0 else x  # se x è maggiore di 0: y=x+2 altrimenti y=x
y
```

```python
# X e Y sono valori booleani
X = True
Y = False

if X or Y:  # basta che uno sia True
    print('Almeno uno è vero')
else:
    print('Entrambi sono falsi')

# Altra sintassi (con continuazione di riga)
if X or \
Y:  # come sopra
    print('Almeno uno è vero')
else:
    print('Entrambi sono falsi')
```

```python
X = True
Y = False

if X and Y:  # entrambi devono essere True
    print('Entrambi sono veri.')
else:
    print('Uno è falso.')

# Altra sintassi (AND bit a bit)
if X & Y:  # come sopra per booleani
    print('Entrambi sono veri.')
else:
    print('Uno è falso.')
```

```python
X = True
Y = False

if not X:  # not X non è True, quindi è False e si va in else
    print('X è falso')
else:
    print('X è vero')

if not Y:  # not Y è True, quindi si entra nell'if
    print('Y è falso')
else:
    print('Y è vero')
```

```python
X = True
Y = False
Z = True
A = False

if (X or Y) and (Z or A):  # puoi anche combinare condizioni
    print('Alcune condizioni sono vere')
```

```python
# Un caso particolare (un po' strano)
# Non serve usarlo per forza, ma se lo trovi in codice altrui saprai cosa significa.

4 and 3, 3 and 4  # Qualsiasi valore diverso da 0, vuoto o None è True
# Se entrambi sono True, l'operatore 'and' restituisce il secondo operando: quindi (3, 4)
```

```python
print(4 or 3, 3 or 4)  # 'or' restituisce il primo valore True (a sinistra)
print([] or {})  # controlla il primo (False), poi il secondo (False): restituisce {}
print(4 or [])  # 4 è True
```

```python
# Match case è molto simile a if/else, ma con una sintassi diversa

x = 10

match x:  # dice al codice di confrontare x
    case 10:  # se x è 10 esegue il blocco seguente
        print('È 10!')
    case _:  # qualsiasi altro valore
        print('Non è 10!')
```

```python
day = 'Monday'

match day:
    case 'Monday' | 'Tuesday' | 'Wednesday' | 'Thursday' | 'Friday':  # '|' significa OR
        print('Giorno feriale')
    case 'Saturday' | 'Sunday':
        print('Weekend')
    case _:
        print('Giorno non valido')
```

```python
# state = 1 oppure
# [1, 2, 3] oppure [0, 2, 3] oppure (1, 2, 3) oppure (0, 2, 3) oppure
# dict(a=1, b=2, c=3) oppure dict(a=0, b=2, c=3) oppure altro

state = [0, 2, 3]

match state:
    case 1 | 2 | 3 as what:  # match di letterali interi, what = 1
        print('or', what)

    case [1, 2, what]:  # match di sequenza (1), what = 3
        print('list', what)
    case [0, *what]:  # match di sequenza (0), what = [2, 3]
        print('list', what)

    case {'a': 1, 'b': 2, 'c': what}:  # match di mapping, what = 3
        print('dict', what)
    case {'a': 0, **what}:  # match di mapping, what = {'b': 2, 'c': 3}
        print('dict', what)

    case (1, 2, what):  # match di tupla: come [1, 2, what]
        print('tuple', what)
    case (0, *what):  # match di tupla: come [0, *what]
        print('tuple', what)

    case _ as what:  # match di qualsiasi altro caso, what = altro
        print('other', what)
```
