```python
# Possiamo leggere riga per riga dal file
open_file = open('myfile.txt', 'r')  # Apre il file per poterlo leggere
print(open_file.readline())  # Legge la prima riga
print(open_file.readline())  # Legge la seconda riga
open_file.close()  # Chiude il file
```

```python
# Scriviamo una lista in un file
a_list = ['apple',
          'banana',
          'orange',
          'cherry',
          'pear']
open_file = open('fruit.txt', 'w')  # Apre il file per scrivere
open_file.writelines(a_list)  # Scrive la lista nel file
open_file.close()  # Chiude il file

# Vediamo cosa è successo
open_file = open('fruit.txt', 'r')  # Apre il file per poterlo leggere
print(open_file.read())  # Legge l'intero file
open_file.close()

# L'output non è molto leggibile: la lista è stata schiacciata in un'unica parola.
```

```python
a_list = ['apple', 'banana', 'orange', 'cherry', 'pear']

# Usa 'with' per assicurarti che il file venga chiuso correttamente dopo scrittura/lettura, ecc.
with open('fruit.txt', 'w') as file:  # sovrascrive il file precedente
    file.write('\n'.join(a_list) + '\n')  # aggiunge automaticamente i caratteri di a capo

# Vediamo cosa è successo
with open('fruit.txt', 'r') as open_file:  # apre il file per poterlo leggere
    print(open_file.read())  # legge l'intero file
```

```python
# Possiamo rileggere il file dentro una lista
with open('fruit.txt', 'r') as open_file:  # apre il file per poterlo leggere
    fruit_list = open_file.readlines()

print(fruit_list)  # aggiunge '\n' in fondo: sta per newline (a capo)

cleaned_fruits = [fruit.strip() for fruit in fruit_list]  # rimuove i '\n'
print(cleaned_fruits)
```
