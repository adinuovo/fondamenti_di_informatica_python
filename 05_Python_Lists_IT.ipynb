# Liste

```python
L0 = []  # lista vuota
L1 = [1, 2, 3]  # lista
L2 = [1, 'A', True]  # anche questa è una lista con diversi tipi di oggetti
L3 = [[1,2,3],
      ['A','B','C'],
      [True,False,True,False]]  # anche questa è una lista, una lista annidata
print(L0)
print(L1)
print(L2)
print(L3)
```

```python
L4 = list(range(10))  # crea un intervallo di numeri dentro una lista
print(L4)
```

```python
L5 = list('hello')
print(L5)
```

```python
# Concatenazione
L6 = L1 + L2
print(L6)
```

```python
# Ripetizione
L7 = L2 * 3
print(L7)
```

```python
# Indici
L8 = ['apple', 'banana', 'cherry']
print(L8[0])
print(L8[1])
print(L8[-1])
```

```python
# Matrice
matrix = [[1,2,3],
          [4,5,6],
          [7,8,9]]
print(matrix[0])
print(matrix[0][1])
print(matrix[2][2])
```

```python
# Modifica degli elementi nelle liste

greeting = ['Hello', 'World!']
greeting[1] = 'Mars!'
print(greeting)
```

```python
# Confrontare le liste
Num1 = [1,2,3,4]
Num2 = [5,6,7,8]
print(Num1 < Num2)
print(Num1 == Num2)
```

```python
3 in Num1  # Il numero 3 è nella lista Num1?
```

```python
# Iterare su una lista
for i in Num1:
    print(i)
```

```python
# Aggiungere elementi a una lista
Num3 = [1,2]
Num3.append(3)
print(Num3)
Num3.extend([4,5])  # estende la lista con più di un elemento
print(Num3)
Num3.insert(3, 'a')  # inserisce in una posizione specifica
print(Num3)
```

```python
# Ricerca
Alph_Num = [1, 2, 3, 'a', 4, 5, 'a']
print(Alph_Num.index('a'))  # indica la posizione di un elemento
print(Alph_Num.index(1))
print(Alph_Num.count('a'))  # indica quante volte compare un elemento
```

```python
# Invertire, ordinare, copiare e svuotare
letters = ['a','j','c','x','e']
letters.reverse()  # inverte l'ordine della lista
print(letters)
letters.sort()  # ordina gli elementi (devono essere dello stesso tipo)
print(letters)
letters_copy = letters.copy()  # copia la lista, altrimenti la nuova variabile resterebbe collegata all'originale
letters_again = letters  # non è una copia: resta collegata a letters

letters.clear()  # svuota la lista: si ottiene una lista vuota []
print(letters)  # è stata svuotata: l'output è []
print(letters_again)  # anche questa risulta svuotata
print(letters_copy)  # non viene svuotata: la copia rimane
```

```python
# Ancora sull'ordinamento

fruit = ['apple','pear','banana','orange']

print(sorted(fruit))
print(sorted(fruit, reverse=True))  # restituisce l'ordinamento inverso
print(sorted(fruit, key=len))  # ordina in base alla lunghezza dell'elemento
print(fruit)  # la lista originale rimane invariata
```

```python
# Usare .sort() modifica la lista
fruit = ['apple','pear','banana','orange']
fruit.sort()
print(fruit)
```

```python
# Ridurre (rimuovere elementi)
Alph_Num = [1, 2, 3, 'a', 4, 5, 'a']
Alph_Num.remove('a')  # rimuove 'a'; Nota: rimuove solo la prima occorrenza
print(Alph_Num)
print(Alph_Num.pop(0))  # rimuove un indice specifico
print(Alph_Num)
```

```python
# Altre eliminazioni
Alph_Num = [1, 2, 3, 'a', 4, 5, 'a']
del Alph_Num[0]
print(Alph_Num)
del Alph_Num[1:3]
print(Alph_Num)
Alph_Num[0:2] = []
print(Alph_Num)
```

```python
# A differenza delle stringhe, puoi modificare un elemento
numbers = [1,2,3,4,5,6,7]
numbers[3] = 50
print(numbers)

numbers[4:8] = [10, 100, 1000]
print(numbers)
```

```python
# Comprensione di lista
mult = [ii * 4 for ii in 'hello']
mult
```

```python
# Ciclo equivalente
mult2 = []
for ii in 'hello':
    mult2.append(ii * 4)
mult2
```
