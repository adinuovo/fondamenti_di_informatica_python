# Funzioni

```python
# Le funzioni seguono un formato specifico

def add_two(number):  # Definisce una funzione: qui il nome della funzione è add_two
    # Tra parentesi c'è un oggetto che puoi passare alla funzione

    plus_two = number + 2  # La funzione esegue l'operazione: aggiunge 2 al numero in input
    return plus_two  # return restituisce l'oggetto che vuoi in output
```

```python
add_two(5)  # Proviamo: inserendo 5 dovrebbe restituire 5+2, cioè 7
```

```python
x = 10.5
add_two(x)
```

```python
# add_two("hello ")

# Questo genera un errore perché la funzione somma un intero a un altro oggetto,
# quindi l'input deve essere un int oppure un float.
```

```python
# Le funzioni sono molto utili per ridurre la quantità di codice da scrivere.
# Inoltre, le funzioni possono essere riutilizzate più volte per scopi diversi.

def square_add_3(number):
    return number * number + 3


num = 2
for i in range(10):
    print(square_add_3(i) + square_add_3(num))

## Invece di:
# num = 2
# for i in range(10):
#     print((i**2+3)+(num**2+3))  # Tutte quelle parentesi possono confondere.
```

```python
# Una funzione non deve per forza avere un input

def print_hello():
    print("hello")

print_hello()
```

## Funzioni lambda

```python
# Un altro modo per definire una funzione è usare lambda

add_two = lambda x: x + 2

add_two(5)
```

```python
# Un vantaggio è che non devi necessariamente dare un nome e possono essere usate dentro altri oggetti.

# Esempio con sorted
names = ['Charlie', 'Alice', 'bob']
print(sorted(names))

sorted_names = sorted(names, key=lambda x: x.lower())  # La funzione in key rende tutto minuscolo; altrimenti le maiuscole verrebbero prima. Vedi il primo output.
print(sorted_names)
```

```python
nums = [3, 4, 2, 1]
sorted(nums, key=lambda x: -x)  # key rende i numeri negativi prima di ordinarli, ma restituisce i numeri originali
```

```python
# Lambda è meglio per funzioni brevi e concise. Per funzioni più complesse, usa def.
```

# Creare e salvare funzioni da importare

```python
## Creiamo prima un file .py: verrà salvato nella directory di lavoro.

# Questo script crea un nuovo file Python e ci scrive una funzione

function_code = """
def greet(name):
    """Restituisce un messaggio di saluto per il nome fornito."""
    return f"Hello, {name}!"
"""

# Scrivi la funzione su file
with open("greetings.py", "w") as file:
    file.write(function_code)

print("Funzione scritta in greetings.py")
```

```python
# Ora dovrebbe esserci un file chiamato greetings.py
# Aprilo e controllalo! Contiene una funzione definita.
# Possiamo chiamare la funzione importando il modulo.

import greetings
```

```python
# Ora possiamo usare la funzione al suo interno.

greetings.greet("Kitty")
```

```python
# Possiamo anche importare solo la funzione
from greetings import greet

greet("Kitty")  # Così non devi scrivere greetings.greet
```

```python
# Puoi anche importare il modulo .py con un nome diverso per digitare più velocemente.
import greetings as g

g.greet("Kitty")
```

```python
# Nota: il file .py può contenere molte funzioni. Potrai poi chiamarle dopo aver importato il modulo.
```
